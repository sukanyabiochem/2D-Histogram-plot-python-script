```python
##Libraries to run the tools##
# libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("channel3_lig-01-30-equil.dat", delim_whitespace=True, names= ['x','angle'])
```

```python
df.angle
```

```python
## use seaborn to create the 2D histogram ##
# set seaborn style
from pylab import*
rc('axes', linewidth=2.5)
fig, ax = plt.subplots(1, 1, figsize=(14, 10))
#plt.rcParams.update({'font.size': 30})
sns.set_style("white")
#cbar_ax=fig.add_axes([0.91,0.125,0.025,0.755])
ax=sns.histplot(x=df.x, y=df.angle, cmap="winter_r", cbar=True, alpha=1, stat='count', bins='auto', binwidth=None, vmin=0, vmax=700)
#ax=sns.histplot(x=df.x, y=df.angle, cmap="winter_r", cbar=True, alpha=1, stat='density', bins='auto', binwidth=None, vmin=0, vmax=0.03)
#res=sns.histplot(x=df.x, y=df.angle, cmap="winter_r", stat='count', bins='auto', binwidth=None, vmin=0, vmax=700)
#cbar=res.collection[0].colorbar

##change the axis lable, fontsize and lable size of the plot## 
plt.xlabel('Ligand position along X-axis', fontsize=30, labelpad=15, fontname='arial')
plt.ylabel('Angle of the ligand (\u03B8)', fontsize=30, labelpad=15, fontname='arial')
#ax.set_title('Channel 3 : ligand conformation', fontsize=32, fontname='arial')
plt.ylim(60, 180, 30)
plt.xlim(-20, 20, 2)
plt.tick_params(axis='both', which='major', labelsize=30, direction='in', length=14, pad=10, bottom='bool', left='bool', width=2)

##change the axis lable, fontsize and lable size of the colorbar## 
ax.figure.axes[-1].set_ylabel('ncounts', size=30, labelpad=10, loc='center')
ax.figure.axes[-1].tick_params(labelsize=28, direction='out', length=14, width=2, pad=10)
ax.get_lines()
#cbar.ax.tick_params(labelsize=40, direction='out', length=20, width=2, pad=10, left='bool')
#cbar.set_ylabel('Contact Frequency', weight='bold', fontsize=50, fontname='arial')
##plt.grid(color = 'red', linestyle = '--', linewidth = 0.3)
##plt.savefig('prob-dis-conformation-1.5us-channel3.png', bbox_inches='tight', dpi=600)
plt.savefig('ch3-github.eps', format='eps', dpi=300)
plt.show()
```
